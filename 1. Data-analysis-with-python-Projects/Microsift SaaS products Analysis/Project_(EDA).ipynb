###Statistical hypothesis Testing

**1. Hypothesis Testing for IsMACCEnrolledProduct and TotalPriceBeforeTaxCDAmount**

- Null Hypothesis (H0): There is no significant difference in revenue between MACC enrolled and non-enrolled products.
- Alternative Hypothesis (H1): MACC enrolled products generate higher revenue compared to non-enrolled products.

```python
# Split the data into two groups based on IsMACCEnrolledProduct
enrolled_prices = merged_df[merged_df['IsMACCEnrolledProduct']]['TotalPriceBeforeTaxCDAmount']
not_enrolled_prices = merged_df[~merged_df['IsMACCEnrolledProduct']]['TotalPriceBeforeTaxCDAmount']

# Perform a t-test to compare the means of the two groups
t_stat, p_value = stats.ttest_ind(enrolled_prices, not_enrolled_prices, equal_var=False)

# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis. There is a significant difference in revenue between MACC enrolled and non-enrolled products.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in revenue between MACC enrolled and non-enrolled products.")
```

**2. Hypothesis Testing for ChannelName and TotalPriceBeforeTaxCDAmount**
- Null Hypothesis (H0): There is no significant difference in revenue among different channels.
- Alternative Hypothesis (H1): Revenue differs significantly among different channels.


```python
# Split the data into groups based on ChannelName
direct_prices = merged_df[merged_df['ChannelName'] == 'Direct']['TotalPriceBeforeTaxCDAmount']
partner_led_prices = merged_df[merged_df['ChannelName'] == 'Partner-Led']['TotalPriceBeforeTaxCDAmount']
field_led_prices = merged_df[merged_df['ChannelName'] == 'Field-Led/EA']['TotalPriceBeforeTaxCDAmount']

# Perform ANOVA test to compare the means of the three groups
f_stat, p_value = stats.f_oneway(direct_prices, partner_led_prices, field_led_prices)

# Print the results
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis. Revenue differs significantly among different channels.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in revenue among different channels.")
```

**3. Hypothesis Testing for ServiceName and TotalPriceBeforeTaxCDAmount**
- Null Hypothesis (H0): There is no significant difference in revenue among different service names.
- Alternative Hypothesis (H1): Revenue differs significantly among different service names.

```python
# Split the data into groups based on ServiceName
vm_license_prices = merged_df[merged_df['ServiceName'] == 'Virtual Machine Licenses']['TotalPriceBeforeTaxCDAmount']
saas_prices = merged_df[merged_df['ServiceName'] == 'SaaS']['TotalPriceBeforeTaxCDAmount']
powerbi_prices = merged_df[merged_df['ServiceName'] == 'Azure PowerBI Visual Service']['TotalPriceBeforeTaxCDAmount']
dynamics_bc_prices = merged_df[merged_df['ServiceName'] == 'Dynamics 365 BC']['TotalPriceBeforeTaxCDAmount']
dynamics_ce_prices = merged_df[merged_df['ServiceName'] == 'Dynamics 365 CE']['TotalPriceBeforeTaxCDAmount']

# Perform ANOVA test to compare the means of the groups
f_stat, p_value = f_oneway(vm_license_prices, saas_prices, powerbi_prices, dynamics_bc_prices, dynamics_ce_prices)

# Print the results
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis. Revenue differs significantly among different service names.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in revenue among different service names.")
```

**4. Hypothesis Testing for ActionName and TotalPriceBeforeTaxCDAmount**
- Null Hypothesis (H0): There is no significant difference in revenue among different action names.
- Alternative Hypothesis (H1): Revenue differs significantly among different action names.

```python
# Split the data into groups based on ActionName
consume_prices = merged_df[merged_df['ActionName'] == 'consume']['TotalPriceBeforeTaxCDAmount']
purchase_prices = merged_df[merged_df['ActionName'] == 'purchase']['TotalPriceBeforeTaxCDAmount']
refund_prices = merged_df[merged_df['ActionName'] == 'refund']['TotalPriceBeforeTaxCDAmount']
renew_prices = merged_df[merged_df['ActionName'] == 'renew']['TotalPriceBeforeTaxCDAmount']

# Perform ANOVA test to compare the means of the groups
f_stat, p_value = f_oneway(consume_prices, purchase_prices, refund_prices, renew_prices)

# Print the results
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis. Revenue differs significantly among different action names.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in revenue among different action names.")
```

**5. Hypothesis Testing for StarRatingMarketplace and TotalPriceBeforeTaxCDAmount**
- Null Hypothesis (H0): There is no significant difference in revenue between products with high and low marketplace ratings.
- Alternative Hypothesis (H1): Products with high marketplace ratings generate higher revenue compared to those with low ratings.

```python
# Split the data into two groups based on StarRatingMarketplace
high_rating_prices = merged_df[merged_df['StarRatingMarketplace'] > 4]['TotalPriceBeforeTaxCDAmount']
low_rating_prices = merged_df[merged_df['StarRatingMarketplace'] <= 4]['TotalPriceBeforeTaxCDAmount']

# Perform a t-test to compare the means of the two groups
t_stat, p_value = stats.ttest_ind(high_rating_prices, low_rating_prices, equal_var=False)

# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis. There is a significant difference in revenue between products with high and low marketplace ratings.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in revenue between products with high and low marketplace ratings.")
```

**6. Hypothesis Testing for NumberOfRatingsExternal and TotalPriceBeforeTaxCDAmount**
